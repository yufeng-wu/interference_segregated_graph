```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../main/autog_results.csv')
df_2000 = pd.read_csv('../result/autog_results/autog_results_2000.csv')
df_5000 = pd.read_csv('../result/autog_results/autog_results_5000.csv')
df_8000 = pd.read_csv('../result/autog_results/autog_results_8000.csv')
```

```python
true_effect = df['True Effect'][0]
df = df.drop(columns=['True Effect'])
```

```python
percentiles = df.apply(lambda x: {
    'mean': np.mean(x),
    '5th_percentile': np.percentile(x, 5),
    '95th_percentile': np.percentile(x, 95)
}, result_type='expand')

percentiles = percentiles.T
```

```python
percentiles
```

```python
n_units = percentiles.index.astype(str).str.extract('(\d+)')[0].astype(int)
means = percentiles['mean']
lower_bounds = percentiles['5th_percentile']
upper_bounds = percentiles['95th_percentile']

plt.figure(figsize=(10, 6))

# Plot the mean values
plt.plot(n_units, means, marker='o', label='Mean values')

# Plot the percentile ranges
plt.fill_between(n_units, lower_bounds, upper_bounds, color='skyblue', alpha=0.5, label='5th to 95th percentile')

# Plot the true effects line
plt.axhline(y=true_effect, color='r', linestyle='-', label='True Effects')

plt.xlabel('Number of Units')
plt.ylabel('Values')
plt.title('True Effects and 95% Mid Percentiles for Each # Units')
plt.legend()
plt.show()
```
