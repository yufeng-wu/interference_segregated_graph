```python
def dgp(n=10000):
    # Generating continuous variables U1 and U2 from a normal distribution
    U1 = np.random.normal(0, 2, n)  # Mean = 0, Std Dev = 1
    U2 = np.random.normal(0, 0.5, n)  # Mean = 0, Std Dev = 1

    # Convert U1, U1+U2, and U2 through expit to get probabilities
    prob_Y1 = expit(2 * U1)
    prob_Y2 = expit(-4 * U1 + 10 * U2)
    prob_Y3 = expit(40 * U2)

    # Generate binary outcomes Y1, Y2, and Y3 using binomial distribution
    Y1 = np.random.binomial(1, prob_Y1, n)
    Y2 = np.random.binomial(1, prob_Y2, n)
    Y3 = np.random.binomial(1, prob_Y3, n)
    
    data = pd.DataFrame({"Y1": Y1, "Y2": Y2, "Y3": Y3})
    
    return data

def test_rf(data, test_set):
    X_Y2 = data[['Y2']]  
    X_Y2Y3 = data[['Y2', 'Y3']]  
    y = data['Y1']  
    
    test_X_Y2 = test_set[['Y2']]  
    test_X_Y2Y3 = test_set[['Y2', 'Y3']]  
    test_y = test_set['Y1']  

    # Predict Y1 using Y2 with Random Forest
    rf_Y2 = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_Y2.fit(X_Y2, y)
    predictions_Y2 = rf_Y2.predict_proba(test_X_Y2)[:, 1]  # Get probability of class 1
    brier_score_Y2 = brier_score_loss(test_y, predictions_Y2)

    # Predict Y1 using Y2 and Y3 with Random Forest
    rf_Y2Y3 = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_Y2Y3.fit(X_Y2Y3, y)
    predictions_Y2Y3 = rf_Y2Y3.predict_proba(test_X_Y2Y3)[:, 1]  # Get probability of class 1
    brier_score_Y2Y3 = brier_score_loss(test_y, predictions_Y2Y3)

    # BIC cannot be directly calculated for Random Forest models as it requires likelihood information
    # Instead, you might compare models based on out-of-bag score, cross-validation, or another suitable metric

    return brier_score_Y2, brier_score_Y2Y3

def test(data, test_set):
    model_Y2 = smf.glm(formula="Y1 ~ Y2", data=data, family=sm.families.Binomial()).fit()
    
    predictions_Y2 = model_Y2.predict(test_set[['Y2']])
    brier_score_Y2 = brier_score_loss(test_set['Y1'], predictions_Y2)

    # Predict Y1 using Y2 and Y3
    model_Y2Y3 = smf.glm(formula="Y1 ~ Y2 + Y3", data=data, family=sm.families.Binomial()).fit()
    
    predictions_Y2Y3 = model_Y2Y3.predict(test_set[['Y2', 'Y3']])
    brier_score_Y2Y3 = brier_score_loss(test_set['Y1'], predictions_Y2Y3)

    model_Y2_bic = -2 * model_Y2.llf + np.log(len(data)) * (len(model_Y2.params))
    model_Y2Y3_bic = -2 * model_Y2Y3.llf + np.log(len(data)) * (len(model_Y2Y3.params))

    return model_Y2Y3_bic < model_Y2_bic, brier_score_Y2, brier_score_Y2Y3

for i in range(100):  # Reduced loop count for brevity
    data = dgp()
    test_set = dgp()
    brier_score_Y2, brier_score_Y2Y3 = test_rf(data, test_set)
    print(brier_score_Y2 > brier_score_Y2Y3, brier_score_Y2, brier_score_Y2Y3)
    # print(result, brier_score_Y2Y3 < brier_score_Y2)

# def test(data):
#     # Predict Y1 using Y2
#     model_Y2 = smf.glm(formula="Y1 ~ Y2", data=data, family=sm.families.Binomial()).fit()

#     # Predict Y1 using Y2 and Y3
#     model_Y2Y3 = smf.glm(formula="Y1 ~ Y2 + Y3", data=data, family=sm.families.Binomial()).fit()
    
#     model_Y2_bic = -2*model_Y2.llf + np.log(len(data))
#     model_Y2Y3_bic = -2*model_Y2Y3.llf + 2*np.log(len(data))
    
#     print(model_Y2Y3_bic, model_Y2_bic)
#     return model_Y2Y3_bic < model_Y2_bic

# for i in range(100):
    
#     data = dgp()
#     print(test(data))
```

```python
# Generating continuous variables U1 and U2 from a normal distribution
U1 = np.random.uniform(0, 1, 10000)  # Mean = 0, Std Dev = 1
U2 = np.random.normal(0, 1, 10000)  # Mean = 0, Std Dev = 1

# Convert U1, U1+U2, and U2 through expit to get probabilities
prob_Y1 = expit(5 * U1)
prob_Y2 = expit(2 * U1 + 3 * U2)
prob_Y3 = expit(4 * U2)

# Generate binary outcomes Y1, Y2, and Y3 using binomial distribution
Y1 = np.random.binomial(1, prob_Y1, 10000)
Y2 = np.random.binomial(1, prob_Y2, 10000)
#Y3 = 2*Y2
Y3 = np.random.binomial(1, prob_Y3, 10000)

# Predict Y1 using Y2
model_Y2 = LogisticRegression(C=1e10)
#model_Y2 = RandomForestClassifier()
model_Y2.fit(Y2.reshape(-1, 1), Y1)

# Predict Y1 using Y2 and Y3
Y2_Y3_combined = np.vstack((Y2, Y3)).T
model_Y2_Y3 = LogisticRegression(C=1e10)
#model_Y2_Y3 = RandomForestClassifier()
model_Y2_Y3.fit(Y2_Y3_combined, Y1)

# Generate test set
U1_test = np.random.uniform(0, 1, 10000)  # Test U1
U2_test = np.random.normal(0, 1, 10000)  # Test U2

# Calculate probabilities for Y1, Y2, Y3 in the test set using expit
prob_Y1_test = expit(5 * U1_test)
prob_Y2_test = expit(2 * U1_test + 3 * U2_test)
prob_Y3_test = expit(4 * U2_test)

# Generate binary outcomes for Y1, Y2, Y3 in the test set
Y1_test = np.random.binomial(1, prob_Y1_test, 10000)
Y2_test = np.random.binomial(1, prob_Y2_test, 10000)
#Y3_test = 
Y3_test = np.random.binomial(1, prob_Y3_test, 10000)

# Predict Y1 using Y2 for the test set
Y1_pred_Y2 = model_Y2.predict(Y2_test.reshape(-1, 1))

# Predict Y1 using Y2 and Y3 for the test set
Y2_Y3_test_combined = np.vstack((Y2_test, Y3_test)).T
Y1_pred_Y2_Y3 = model_Y2_Y3.predict(Y2_Y3_test_combined)

# Calculate accuracy for both models
accuracy_Y2 = accuracy_score(Y1_test, Y1_pred_Y2)
accuracy_Y2_Y3 = accuracy_score(Y1_test, Y1_pred_Y2_Y3)

print("Y1 using Y2", np.mean((Y1_test - model_Y2.predict_proba(Y2_test.reshape(-1, 1))[:, 1])**2))
print("Y1 using Y2, Y3", np.mean((Y1_test - model_Y2_Y3.predict_proba(Y2_Y3_test_combined)[:, 1])**2))
#print("Accuracy for model predicting Y1 using Y2:", accuracy_Y2)
#print("Accuracy for model predicting Y1 using Y2 and Y3:", accuracy_Y2_Y3)


#print(model_Y2.coef_, model_Y2_Y3.coef_)
```

```python
# Generating continuous variables U1 and U2 from a normal distribution
U1 = np.random.normal(0, 1, 10000)  
U2 = np.random.normal(0, 1, 10000)  

# Generating continuous outcomes for Y1, Y2, and Y3
Y1 = 4 * U1 + np.random.normal(0, 0.1, 10000)
Y2 = 3 * U1 + 10 * U2 + np.random.normal(0, 0.1, 10000)
Y3 = 2 * U2 + np.random.normal(0, 0.1, 10000)

# Linear Regression to predict Y1 using Y2
model_Y2 = LinearRegression()
model_Y2.fit(Y2.reshape(-1, 1), Y1)

# Linear Regression to predict Y1 using Y2 and Y3
Y2_Y3_combined = np.vstack((Y2, Y3)).T
model_Y2_Y3 = LinearRegression()
model_Y2_Y3.fit(Y2_Y3_combined, Y1)

# Generate test set for continuous outcomes
U1_test = np.random.normal(0, 1, 10000)
U2_test = np.random.normal(0, 1, 10000)
Y1_test = 4 * U1_test + np.random.normal(0, 0.1, 10000)
Y2_test = 3 * U1_test + 10 * U2_test + np.random.normal(0, 0.1, 10000)
Y3_test = 2 * U2_test + np.random.normal(0, 0.1, 10000)

# Predict Y1 using Y2 for the test set
Y1_pred_Y2 = model_Y2.predict(Y2_test.reshape(-1, 1))

# Predict Y1 using Y2 and Y3 for the test set
Y1_pred_Y2_Y3 = model_Y2_Y3.predict(np.vstack((Y2_test, Y3_test)).T)

# Calculate mean squared error for both models
mse_Y2 = mean_squared_error(Y1_test, Y1_pred_Y2)
mse_Y2_Y3 = mean_squared_error(Y1_test, Y1_pred_Y2_Y3)

print("MSE for model predicting Y1 using Y2:", mse_Y2)
print("MSE for model predicting Y1 using Y2 and Y3:", mse_Y2_Y3)
```
